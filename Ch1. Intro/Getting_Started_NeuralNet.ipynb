```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(
    style='darkgrid', 
    rc={'axes.facecolor': '.9', 'grid.color': '.8'}
)
sns.set_palette(palette='deep')
%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 6]
plt.rcParams['figure.dpi'] = 100
```

```python
# Number of samples (training set). 
n = 500
x = np.linspace(start=0, stop=1, num=n)

# x, y는 학습 데이터
def f(x):
    f = np.sin((4*np.pi)*x) + np.sin((7*np.pi)*x)
    return(f)
```

```python
# Error standard deviation. 
sigma_n = 0.4
# Errors.
epsilon = np.random.normal(loc=0, scale=sigma_n, size=n)

# Observed target variable. 
f_x = f(x)
y = f_x + epsilon
```

```python
# Plotting with seaborn library
fig, ax = plt.subplots()
# Plot training data.
sns.scatterplot(x=x, y=y, label='training data', ax=ax);
# Plot "true" linear fit.
sns.lineplot(x=x, y=f_x, color='blue', label='f(x)', ax=ax);

ax.set(title='Sample Data')
ax.legend(loc='upper right');
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import math
```

```python
# Create the model 
model = keras.Sequential()
model.add(keras.layers.Dense(units = 1, activation = 'linear', input_shape=[1]))
model.add(keras.layers.Dense(units = 64, activation = 'elu'))
model.add(keras.layers.Dense(units = 64, activation = 'elu'))
model.add(keras.layers.Dense(units = 1, activation = 'linear'))
model.compile(loss='mse', optimizer="adam")

model2 = keras.Sequential()
model2.add(keras.layers.Dense(units = 1, activation = 'linear', input_shape=[1]))
model2.add(keras.layers.Dense(units = 64, activation = 'elu'))
model2.add(keras.layers.Dense(units = 64, activation = 'elu'))
model2.add(keras.layers.Dense(units = 1, activation = 'linear'))
model2.compile(loss='mse', optimizer="adam")

# Display the model
model.summary()
```

```python
# Training 1
model.fit(x, y, epochs=5000, verbose=1)
```

```python
# Training 2
model2.fit(x, y, epochs=500, verbose=1)
```

```python
# Compute the output 
y1_predict = model.predict(x)
y2_predict = model2.predict(x)
print(x.shape, y2_predict[:,0].shape, f_x.shape)
```

```python
################ Plot #################
fig, ax = plt.subplots()
# Plot training data.
A = sns.scatterplot(x=x, y=y, label='training data', ax=ax)
# Plot "true" linear fit.
B = sns.lineplot(x=x, y=f_x, color='blue', label='base func.', ax=ax);
# Plot prediction. 
C = sns.lineplot(x=x, y=y2_predict[:,0], color='green', label='pred 500', linestyle = ':', ax=ax, linewidth = 2.2)
D = sns.lineplot(x=x, y=y1_predict[:,0], color='red', label='pred 5000', linestyle = '--', ax=ax)

handles, labels = ax.get_legend_handles_labels() ## 범례 처리되는 요소와 해당 라벨 
dict_labels_handles = dict(zip(labels, handles))
labels = ['training data','base func.','pred 500','pred 5000'] ## 원하는 순서 라벨
handles = [dict_labels_handles[l] for l in labels] ## 라벨 순서에 맞게 요소 재배치
ax.set_title('ANN Model Prediction vs. Training Dataset')
ax.legend(handles,labels,loc='upper right');
```
